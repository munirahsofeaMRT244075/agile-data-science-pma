```python
import pandas as pd

url = "https://raw.githubusercontent.com/munirahsofeaMRT244075/agile-data-science-pma/refs/heads/main/data./telco.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
print(len(df))
print(len(df.columns))
print(df.columns.tolist())
```

Q1 (B)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['MonthlyCharges'], bins=30, kde=True)
plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='tenure', y='MonthlyCharges', hue='Churn', alpha=0.5)
plt.title("Tenure vs Monthly Charges, coloured by Churn")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title("Churn Count by Contract Type")
plt.show()
```
